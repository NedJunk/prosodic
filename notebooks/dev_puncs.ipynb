```python
def _to_html_df(self, parse=None, as_str=False, css=HTML_CSS, blockquote=False):
    if parse is None: parse=self.best_parse
    
    output = []

    for wordtoken in self.wordtokens:
        prefstr = get_initial_whitespace(wordtoken.txt)
        if prefstr:
            odx = {'txt':prefstr}
            output.append(odx)

        wordtoken_slots = parse.wordtoken2slots[wordtoken]
        if wordtoken_slots:
            for slot in wordtoken_slots:
                pos = slot.parent
                spclass = 's' if slot.is_prom else 'w'
                stclass = 's' if slot.unit.is_stressed else 'w'
                vclass = 'y' if pos.violset else 'n'

                odx = {'txt':slot.unit.txt, 'meter':spclass, 'stress':stclass, 'viol':vclass}
                output.append(odx)
        else:
            odx = {'txt':wordtoken.txt}
            output.append(odx)
    

    return pd.DataFrame(output)
```

```python
df=_to_html_df(line)
df.fillna(method='ffill').head(10)
```
